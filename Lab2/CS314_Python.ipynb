**Procedural** (from http://rosettacode.org/wiki/Sorting_algorithms/Insertion_sort#Python)

```python
def insertion_sort(l):
    for i in xrange(1, len(l)):
        j = i-1 
        key = l[i]
        while (l[j] > key) and (j >= 0):
           l[j+1] = l[j]
           j -= 1
        l[j+1] = key
```

```python
nums = list(randint(0,100,10))
print nums
insertion_sort(nums)
print nums
```

**Object Oriented**

```python
class Sphere:
    '''A sphere at the origin of the coordinate system.'''
    
    template = 'sphere( r = {radius}, $fa = {fa}, $fs = {fs}, $fn = {fn} );'
    
    def __init__(self, radius, fa=12, fs=2, fn=0):
        '''Initialize a sphere to the given values.  The special
        parameters fa, fs and fn have default values the same as
        in OpenSCAD'''
        # store parameters in a dictionary
        self.a = 5
        self.params = dict()
        self.params['radius'] = radius
        self.params['fa']     = fa          # Minimum angle in degrees for a fragment
        self.params['fs']     = fs          # Angle in mm
        self.params['fn']     = fn          # Resolution
        
    def __str__(self):
        ''' Return a string representation of this object.
            The equivalent of toString() in Java
        '''
        return Sphere.template.format(**self.params)
```

```python
s1 = Sphere(5)
```

```python
# a generator that yields items instead of returning a list
def firstn(n):
    num = 0
    while num < n:
        yield num
        num += 1

sum_of_first_n = sum(firstn(1000000))
```

```python
def fib():
    a, b = 0, 1
    while 1:
        yield b
        a, b = b, a+b
```

*Context Managers* (python's with statement)

```python
#!/usr/bin/env python
#
# Python Timer Class - Context Manager for Timing Code Blocks
# Corey Goldberg - 2012
# http://python.dzone.com/articles/python-timer-class-context
#
from timeit import default_timer

class Timer(object):
    def __init__(self, verbose=False):
        self.verbose = verbose
        self.timer = default_timer
        
    def __enter__(self):
        self.start = self.timer()
        return self
        
    def __exit__(self, *args):
        end = self.timer()
        self.elapsed_secs = end - self.start
        self.elapsed = self.elapsed_secs * 1000 # millisecs
        if self.verbose:
            print 'elapsed time: %f ms' % self.elapsed

def run_it(url):
    # example:
    # 'HTTP GET' from requests module, inside timer blocks.
    # invoke the Timer context manager using the `with` statement.
    
    import requests
    
    # verbose (auto) timer output
    with Timer(verbose=True):
        r = requests.get(url)
    
    # print stored elapsed time in milliseconds
    with Timer() as t:
        r = requests.get(url)
    print 'response time (millisecs): %.2f' % t.elapsed
    
    # print stored elapsed time in seconds
    with Timer() as t:
        r = requests.get(url)
    print 'response time (secs): %.3f' % t.elapsed_secs
```

```python
run_it('https://github.com/timeline.json')
```

```python
run_it('http://www.wou.edu')
```

### Operator Overloading

```python
# Not new or different, but not available in Java
# for methods to overload: http://rgruet.free.fr/PQR27/PQR2.7.html#SpecialMethods
class Box:
    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z
    def __add__(self,rhs):
        if type(rhs) is int:
            return Box(self.x + rhs, self.y + rhs, self.z + rhs)
        else:
            newx = self.x + rhs.x
            newy = self.y + rhs.y
            newz = self.z + rhs.z
            return Box(newx,newy,newz)
    def __str__(self):
        return 'Box({},{},{})'.format(self.x,self.y,self.z)
```

```python
b1 = Box(1,2,3)
b2 = Box(10,20,30)
print b1 + 9
print b1 + b2
```

### Make your own iterable object

```python
# From 'Generator Tricks for Systems Programmers' http://www.dabeaz.com
class Countdown(object):
    def __init__(self,start):
        self.count = start
    def __iter__(self):
        return self
    def next(self):
        if self.count <= 0:
            raise StopIteration
        r = self.count
        self.count -= 1
        return r
```

```python
c = Countdown(5)
for i in c:
    print i
```
